```python
import pandas as pd
import numpy as np
movies=pd.read_csv('movies.csv',index_col=0)
directors=pd.read_csv('directors.csv',index_col=0)
```

How can we check all the director_id value is present in id od directors??


```python
np.all(movies['director_id'].isin(directors['id'])) # so the are present in director id
```

```python
#now we are merging two datasets of 'movies' and 'directors' for further data analysis
data=movies.merge(directors,how='left',left_on='director_id',right_on='id')
```

```python
data
```

```python
#now we don't need director_id and id_y columns so we are dropping it permanently
data.drop(['director_id','id_y'],axis=1,inplace=True)
```

```python
data
```

```python
#now we will convert gender column into 0 and 1 by 'apply' method
Male=0
Female=1
def encode(gender):
  if gender=='Male':
    return 0
  else:
    return 1

```

```python
data['gender']=data['gender'].apply(encode)
```

```python
data
```

```python
#now we will find out the total summation of revenue and budget by using apply method
data[['revenue','budget']].apply(np.sum) # it is taking every row and calculating the sum of rev and budget seperately
```

```python
#now we need sum of revenue and budget for every row
data[['revenue','budget']].apply(np.sum,axis=1)
```

```python
#find out the profit
def prof(df):
  profit=df['revenue']-df['budget']
  return profit
```

```python
data['profit']=data[['revenue','budget']].apply(prof,axis=1)
```

```python
data
```

```python
#groupby
#find out the count of movies for every director
data.groupby('director_name')['title'].count()
```

```python
#find out the year of last movie for every director
data.groupby('director_name')['year'].max()
```

```python
#when we need more than 1 value we use aggregate function
#finding the year of first and last movie for every director
data.groupby('director_name')['year'].aggregate(['min','max'])
```

```python
#group-base filtering
#find the details of the movies by the high budget directors(budget>=100M)
high_budget_directors=data.groupby('director_name').filter(lambda df:df['budget'].max()>=100000000)
```

```python
high_budget_directors
```

```python
#Filtering Risky Movie
def func(df):
  df['risky']=df['budget']-df['revenue'].mean()>=0
  return df
```

```python
data_risky=data.groupby('director_name').apply(func)
data_risky[data_risky['risky']==True]
```

```python
#Findout the most productive director
data_agg=data.groupby('director_name')[['year','title']].aggregate({'year':['min','max'],'title':'count'})
```

```python
data_agg
```

```python
data_agg.columns
```

```python
new_cols=[]
for i in data_agg.columns:
  new_cols.append('_'.join(i))
new_cols
```

```python
data_agg.columns=new_cols
```

```python
data_agg
```

```python
#now we have to find out the active years
data_agg['active_years']=data_agg['year_max']-data_agg['year_min']
```

```python
data_agg
```

```python
#now we will find out movie per year
data_agg['movie_per_year']=data_agg['title_count']/data_agg['active_years']
```

```python
data_agg.sort_values('movie_per_year',ascending=False).head(3)
```
