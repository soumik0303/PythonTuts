```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data
```

#What is the distribution of passenger ages of the titanic????

```python
plt.hist(data['Age'].dropna())
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

So maximum number of passengers in titanic belonged to the age group of 20 to 30.

#How is the gender distribution among the passengers of Titanic??

```python
sns.countplot(x='Sex',data=data,color='red')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

So,number of male passengers were higher than the female passengers

#what is the distrubution of passenger classes in Titanic???

```python
sns.countplot(x='Pclass',data=data,color='green')
plt.title('Distribution of Passenger_Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

So,class-3 passengers were maximum in number and Class-2 passengers were lowest in numbers

#How does the survival rate varies from passenger class??

```python
sns.countplot(x='Pclass',hue='Survived',data=data,palette='pastel')
plt.title('Survival rate vs passenger class')
plt.xlabel('Pclass')
plt.ylabel('survived')
plt.show()
```

So,survival rate is maximum among class-1 passengers,on the contrary maximum number of annihilliated people were from class-3

#What is the correlation between the Age and fare paid by the passengers??

```python
sns.scatterplot(x='Age',y='Fare',data=data,color='coral')
plt.title('Correlation between age and fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

The correlation between age and fare is concentrated,but in some cases fare is high for the people containing the age between 30 to 40

#How does the survival rate varies based on Gender???


```python
sns.countplot(x='Sex',hue='Survived',data=data,palette='Set2')
plt.title('Gender based survival')
plt.xlabel('Gender')
plt.ylabel=('Survived')
plt.legend(title='Survived',loc='upper right',labels=['No','Yes'])
plt.show()
```

If we analys the survival rate,we will find that the survival rate of women is higher than the men

How does passenger age fare and class contribute to the survival status????


```python
survived_data0=data[data['Survived']==0]
survived_data1=data[data['Survived']==1]
```

```python
sns.scatterpl(x='Age',y='Fare',hue='Pclass',data=survived_data0)
plt.title('Passenger age,fare and class for those who didnt survive')
plt.xlabel=('Age')
plt.legend(title='Passenger_Class',loc='upper right')

plt.show()
```

So,according to the analysis the people from the age group of 20 to 40 those who belonged to the class 3 it means paid less fare than other two classes died more

```python
sns.scatterplot(x='Age',y='Fare',hue='Pclass',data=survived_data1)
plt.title('Passenger age fare and class for those who survived')
plt.legend(title='Passenger_Class')
plt.show()
```

According to the analysis maximum people those who survived belonged to the class 1 group and aged between 20 to 60

#what is relationship among age fare and no of spouse/sibling???

```python
sns.scatterplot(x='Age',y='Fare',size=data['SibSp'],data=data)
plt.title('Relationship among fare,age and no. of company' )
plt.legend(title='Company')
plt.show()
```

Maximum people those those who paid less and belonged to the age gruoup between 10 to 50 had no of company 0 or 1

#How does passenger  age,fare and gender contribute to the survival status????

```python
sns.scatterplot(x='Age',y='Fare',hue='Sex',data=survived_data0)
plt.title('Age Fare and Gender for those who didnt survive')
plt.show()
```

according to the analysis maximum number of men whose who paid less fare died more than the women.

```python
sns.scatterplot(x='Age',y='Fare',hue='Sex',data=survived_data1)
plt.title('Age,Fare and Gender for those who survived')
plt.show()
```

According to the analysis,maximum no of women aged between 0 to 40 survived more and paid less fare than men
