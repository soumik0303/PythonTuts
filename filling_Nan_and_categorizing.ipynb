```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Pfizer_1.csv')
```

```python
data.head()
```

```python
data_melt=pd.melt(data,id_vars=['Date','Drug_Name','Parameter'],var_name='Time',value_name='Reading')
```

```python
data_melt.head()
```

```python
data_tidy=data_melt.pivot(index=['Date','Time','Drug_Name'],columns='Parameter',values='Reading')
```

```python
data_tidy.reset_index(inplace=True)
```

```python
data_tidy
```

```python
def temp_mean(df):
  df['Temperature_avg']=df['Temperature'].mean()
  return df
```

```python
data_tidy=data_tidy.groupby('Drug_Name').apply(temp_mean)
```

```python
data_tidy
```

```python
data_tidy['Temperature'].fillna(data_tidy['Temperature_avg'],inplace=True)
```

```python
data_tidy.head(25)
```

```python
def pres_mean(df):
  df['Pressure_avg']=df['Pressure'].mean()
  return df
```

```python
data_tidy=data_tidy.groupby('Drug_Name').apply(pres_mean)
```

```python
data_tidy
```

```python
data_tidy['Pressure'].fillna(data_tidy['Pressure_avg'],inplace=True)
```

```python
data_tidy.head(15)
```

```python
#pd.cut
data_tidy['Temperature'].min(),data_tidy['Temperature'].max()
```

```python
buckets=[5,20,35,50,60]
bucket_labels=['low','medium','high','very high']
```

```python
data_tidy['temp_cat']=pd.cut(data_tidy['Temperature'],bins=buckets,labels=bucket_labels)
```

```python
data_tidy.head()
```

```python
data_tidy['Pressure'].min(),data_tidy['Pressure'].max()
```

```python
buckets=[1,10,20,30]
bucket_labels=['low','medium','high']
```

```python
data_tidy['press_cat']=pd.cut(data_tidy['Pressure'],bins=buckets,labels=bucket_labels)
```
