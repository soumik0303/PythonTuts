```python
A=[[4,7,9],[7,6,2],[5,1,6]]
```

```python
A
```

```python
#Sum of rows
ans=[]
for i in range(len(A)):
  sum=0
  for j in range(len(A[0])):
    sum=sum+A[i][j]
  ans.append(sum)
print(ans)
```

```python
#Sum of columns
ans=[]
for i in range(len(A[0])):
  sum=0
  for j in range(len(A)):
    sum=sum+A[j][i]
  ans.append(sum)
print(ans)
```

```python
#diagonal sums
sum=0
for i in range(len(A)):
  sum=sum+A[i][i]
print(sum)
```

```python
#largest element in each row
ans=[0]*len(A)
for i in range(len(A)):
  for j in range(len(A[0])):
    ans[i]=max(ans[i],A[i][j])
print(ans)
```
