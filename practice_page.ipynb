```python
matrix1=[[1,2,3],[4,5,6]]
matrix2=[[10,11],[20,21],[30,31]]
result=[[0,0],[0,0]]
if len(matrix1[0])==len(matrix2):
  for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
      for k in range(len(matrix1[0])):
        result[i][j]+=matrix1[i][k]*matrix2[k][j]
  print(result)
else:
  print("not valid")
```
